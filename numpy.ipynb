### Some important Numpy array creation functions
| Functions | Description |
| ------ | ------ |
|`array`|**Convert input data(list,tuple,array or other sequence type) to an ndarray either by inferring a data type or explicity specifying a data type; copies the input data by default**|
|`asarry`|Convert input to ndarray, but do not copy if the input is already an ndarray|
|`arange`|**Like the built-in range but return an ndarry instead of list**|
|`ones,ones_like`|Produce an array of all 1s with the given shape and data type; ones_like takes another array and produces a ones array of the same shape and data type|
|`zeros, zeros_like`|**Like ones and ones_like producing 0 instead**|
|`empty, empty_like`|Create new arrays by allocating new memory, but do not populate with any values like ones and zeros|
|`full, full_like`|**Produce an array of the given shape and data type with all values set to the indicated "fill value"; full_like takes another array and produces a filled array of the same shape and data type**|
|`eye, identy`|Create a square N x N identity matrix (1’s on the diagonal and 0’s elsewhere)|

`To create a higher dimensional array with these methods, pass a tuple for the shape`

````python
# Boolean Indexing
names = np.array(["Bob", "Joe", "Will", "Bob", "Will", "Joe", "Joe"])
data = np.array([[4, 7], [0, 2], [-5, 6], [0, 0], [1, 2],[-12, -4], [3, 4]])
````

````python
names
````

````python
data
````

````python
names == "Bob"
````

````python
#The Boolean array must be of the same length as the array axis it’s indexing. You can even mix and match Boolean arrays with slices or integers (or sequences of integers; more on this later).
data[names == "Bob"]
````

````python
data[names == "Bob", 1:]
````

````python
#select from the rows where names == "Bob" and index the columns, too:
data[names == "Bob", 1]
````

````python
names != "Bob"
````

````python
~(names == "Bob")
````

````python
data[~(names == "Bob")]
````

````python
cond = names == "Bob"
````

````python
data[~cond]
````

````python
names
````

````python
mask = (names == "Bob") | (names == "Will")
mask
````

````python
data[mask]
````

# **`The Python keywords "and" and "or" do not work with Boolean arrays. Use & (and) and | (or) instead.`**

````python
data[data < 0] = 0
data
````

````python
data[names != "Joe"] = 7
data
````

# Fancy **Indexing**

*Fancy indexing is a term adopted by NumPy to describe indexing using integer arrays*



````python
arr = np.zeros((8, 4))
arr
````

````python
for i in range(8):
  arr[i] = i
````

````python
arr
````

````python
#To select a subset of the rows in a particular order, you can simply pass a list or ndarray of integers specifying the desired order:
arr[[4, 3, 0, 6,7]]
````

````python
#Hopefully this code did what you expected! Using negative indices selects rows from the end:
arr[[-3, -5, -7]]
````

````python
'''
Passing multiple index arrays does something slightly different;
it selects a one-dimensional array of elements corresponding to each tuple of indices:
'''
arr = np.arange(32).reshape((8, 4))
````

````python
arr
````

````python
print(arr[5][3])
````

````python
arr[[1, 5, 7, 2], [0, 3, 1, 2]]
````

````python
arr[[1, 5, 7, 2]][:]
````

````python
arr[[1, 5, 7, 2]][:, [0, 3, 1, 2]]
````

# **Pseudorandom Number Generation**
The `numpy.random` module supplements the built-in Python random module with functions for efficiently generating whole arrays of sample values from many kinds of probability distributions. For example, you can get a 4 × 4 array of samples from the standard normal distribution using `numpy.random.standard_normal`:









````python
samples = np.random.standard_normal(size=(4, 4))
samples
````

````python
np.mean(samples)     
````
